### Комментарий к решению
Я предполагал, что их посчитаете каждого по-отдельности, но слово "в совокупности" действительно смущает :D

```python
total_number = int(data.loc[:,['occupation']][((data.sex == 'Мужской') & (data.occupation == 'Врач'))|((data.sex == 'Женский')&(data.occupation == 'Ученый'))|(data.occupation == 'На пенсии')].count())
print('мужчины-врачи, женщины-ученые и пенсионеры', total_number)
```

### Комментарий к решению
Отлично

```python
male_ratings_sum = int(data.loc[:,['impressions_whole']][data.sex=='Мужской'].sum())
male_number = int(data.loc[:,['sex']][data.sex=='Мужской'].count())
male_average = male_ratings_sum/male_number
print('средняя оценка мужчин',male_average)
female_ratings_sum = int(data.loc[:,['impressions_whole']][data.sex=='Женский'].sum())
female_number = int(data.loc[:,['sex']][data.sex=='Женский'].count())
female_average = female_ratings_sum/female_number
print('средняя оценка женщин',female_average)
print('мужчины выше оценивают посещение -',male_average>female_average)
```
