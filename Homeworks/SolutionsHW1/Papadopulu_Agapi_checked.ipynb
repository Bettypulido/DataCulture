```python
data2 = data.drop(['recorddate_key', 'birthdate_key', 'orighiredate_key', 'terminationdate_key', 'city_name', 'department_name', 'job_title', 'store_name', 'gender_short', 'gender_full', 'termreason_desc', 'termtype_desc', 'STATUS_YEAR', 'BUSINESS_UNIT', 'STATUS'], axis=1)
data3 = data2.sort_values(by = 'length_of_service', ascending = False)
data3[0:2]
```

```python
data.pivot_table(values= 'age', index='length_of_service', columns='gender_short', aggfunc='count').plot(kind='bar', stacked=True)
plt.title('Количество проработанных лет')
plt.show()
```
