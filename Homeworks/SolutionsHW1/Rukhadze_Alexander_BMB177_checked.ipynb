# Задание 6.

### Комментарий к решению

Сгруппируем данные по переменной EducationField и проанализируем их по среднему месячному доходу

```python
education_and_income = data.groupby('EducationField').MonthlyIncome.mean()
education_and_income
```

Теперь представим полученные данные на одной столбиковой диаграмме

```python
education_and_income.plot.bar();
```

Исходя из данных и диаграммы, получается, что выгоднее всего было учиться на Маркетинге или Управлении человеческими ресурсами, чтобы, в среднем, получать более высокую заработную плату. Меньше всего повезло тем, кто получил Technical Degree.
