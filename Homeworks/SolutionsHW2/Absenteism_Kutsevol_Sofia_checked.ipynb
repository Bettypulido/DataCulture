````python
from sklearn.model_selection import train_test_split
data_train, data_test = train_test_split(data, test_size = 0.3)

y_train = data_train['time in hours']
y_test = data_test['time in hours']

X_train = data_train.drop(['time in hours'], axis=1)
X_test = data_test.drop(['time in hours'], axis=1)

````

### Комментарий
Шкалировать категориальные переменные точно не стоит, только непрерывные

````python
from sklearn.preprocessing import StandardScaler

# объявили скалировщик!
scaler = StandardScaler()

# учим скалировщик скалировать все переменны на трэйне
scaler.fit(data_train[['Month of absence', 'Day of the week', 
                       'Seasons', 'Disciplinary failure', 'Education',
                       'Social drinker','Social smoker','Pet']]) 
````

````python
data_train_scale = scaler.transform(data_train[['Month of absence', 'Day of the week', 'Seasons', 'Disciplinary failure', 'Education','Social drinker','Social smoker','Pet']])


data_test_scale = scaler.transform(data_test[['Month of absence', 'Day of the week', 'Seasons', 'Disciplinary failure', 'Education','Social drinker','Social smoker','Pet']])

````

````python
data_train[['Month of absence', 'Day of the week', 'Seasons', 'Disciplinary failure', 
            'Education','Social drinker','Social smoker','Pet']] = data_train_scale

data_test[['Month of absence', 'Day of the week', 'Seasons', 'Disciplinary failure', 
           'Education','Social drinker','Social smoker','Pet']] = data_test_scale
````

````python
y_train = data_train['time in hours'] 
y_test = data_test['time in hours']

X_train = data_train.drop('time in hours', axis=1).get_values()
X_test = data_test.drop('time in hours', axis=1).get_values()

````

````python
y_mean = np.mean(y_train)                     
y_pred_naive = np.ones(len(y_test)) * y_mean  
y_pred_naive[:5]
````

````python
from sklearn import metrics
````

````python

metrics.mean_absolute_error(np.exp(y_test), np.exp(y_pred_naive))
````

````python
metrics.mean_squared_error(np.exp(y_test), np.exp(y_pred_naive))
````

````python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mean_absolute_percentage_error(y_test, y_pred_naive)
````

````python
metrics.r2_score(y_test, y_pred_naive)
````

````python
def print_metrics(y_test,y_pred):
    print('MAE:', metrics.mean_absolute_error(np.exp(y_test), np.exp(y_pred)))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(np.exp(y_test), np.exp(y_pred))))
    print('R2:',  metrics.r2_score(y_test, y_pred))
    print('MAPE:', mean_absolute_percentage_error(y_test, y_pred))
    pass
````

````python
print_metrics(y_test, y_pred_naive)
````

````python
from sklearn.linear_model import LinearRegression


model_regression = LinearRegression()


model_regression.fit(X_train, y_train)


y_pred_regr = model_regression.predict(X_test)
````

````python
print_metrics(y_test,y_pred_regr)
````

````python
featureImportance = pd.DataFrame({"feature": data.drop('time in hours',axis=1).columns, 
                                  "importance": model_regression.coef_})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].plot('bar');
````

````python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100)


rf.fit(X_train, y_train)


y_pred_forest = rf.predict(X_test)
````

````python

featureImportance = pd.DataFrame({"feature": data.drop('time in hours',axis=1).columns, 
                                  "importance": rf.feature_importances_})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].plot('bar');
````
