```python
import pandas as pd              
import numpy as np                
import matplotlib.pyplot as plt  
import seaborn as sns            

%matplotlib inline
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_vectorizer = TfidfVectorizer(ngram_range=(1, 2), min_df=0.001, norm='l2', stop_words='english')
```

```python
data['reviews.text'] = data['reviews.text'].str.lower()

X = data['reviews.text']
y = data['reviews.rating']
```

```python
from sklearn.model_selection import train_test_split
# df_train, df_test = train_test_split(df, test_size = 0.3)
```

```python
X = tf_vectorizer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.30)
```

```python
print(X_train.shape) # Посмотрим на размеры трэйна и теста 
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_mean = np.mean(y_train)               
y_pred_naive = np.ones(len(y_test)) * y_mean  
y_pred_naive[:5]
```

```python
from sklearn import metrics  
```

```python
metrics.mean_absolute_error(y_test, y_pred_naive)
```

```python
metrics.mean_absolute_error(np.exp(y_test), np.exp(y_pred_naive))
```

```python
metrics.mean_squared_error(y_test, y_pred_naive)
```

```python
metrics.mean_squared_error(np.exp(y_test), np.exp(y_pred_naive))
```

```python
np.sqrt(metrics.mean_squared_error(np.exp(y_test), np.exp(y_pred_naive)))
```

```python
metrics.r2_score(y_test, y_pred_naive)
```

```python
def print_metrics(y_test,y_pred):
    print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print('R2:',  metrics.r2_score(y_test, y_pred))
    pass
```

```python
print_metrics(y_test, y_pred_naive)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
# Сделали по лесу прогнозы 
y_pred_forest = rf.predict(X_test)
```

```python
print_metrics(y_test, y_pred_forest)
```

```python
featureImportance = pd.DataFrame({"feature": tf_vectorizer.get_feature_names(), 
                                  "importance": rf.feature_importances_})
featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].head(30).plot('bar', figsize=(12,5))
```

```python
from sklearn.linear_model import Ridge

linreg = Ridge()

linreg.fit(X_train, y_train)

y_pred = linreg.predict(X_test)
```

```python
print_metrics(y_test, y_pred)
```

```python
featureImportance = pd.DataFrame({"feature": tf_vectorizer.get_feature_names(), 
                                  "importance": linreg.coef_})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
```

```python
def plotImportance(featureImportance, top_words=20):
    """
        Plot importance of words in linear model
        
        featureImportance: pd.DataFrame
            data with coefs values
        
        top_words: int
            how many words to plot on both negative and positive sides
    """
    featureImportance = featureImportance.head(top_words).append(featureImportance.tail(top_words))
    x = np.arange(len(featureImportance))
    y = featureImportance.importance

    mask1 = y < 0
    mask2 = y >= 0

    plt.figure(figsize=(18, 8))
    plt.bar(x[mask1], y[mask1], color = 'red')
    plt.bar(x[mask2], y[mask2], color = 'blue')
    plt.xticks(np.arange(len(featureImportance)), 
                   featureImportance.index, rotation=90, ha="right", fontsize=11)
    plt.show()
```

```python
plotImportance(featureImportance)
```
