```python
# Из datetime вытащим час матча
data.Datetime = pd.to_datetime(data.Datetime)
data['Hour'] = data.Datetime.apply(lambda x: x.hour)


dummy_hour = pd.get_dummies(data.Hour, drop_first=True)
data = pd.concat([data, dummy_hour], axis=1)

data.drop(['Datetime', 'Hour'], axis=1, inplace=True)
```

Неудивительно, что модельки не работали, ты разбиение на train-test сделал в самом начале, получив переменные для икса и игрека, а только потом стал работать с датасетом, выходит, что в иксе у тебя оставались сырые, необработанные данные

```python
from sklearn.model_selection import train_test_split
data_train, data_test = train_test_split(data, test_size = 0.3)

y_train = data_train['Home Team Win']
y_test = data_test['Home Team Win']

X_train = data_train.drop(['Home Team Win'], axis=1)
X_test = data_test.drop(['Home Team Win'], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression 
# объявляем модель 
logit = LogisticRegression() 
# обучаем на тренировочной выборке 
logit.fit(X_train, y_train) 
y_predicted = logit.predict(X_test)
y_proba = logit.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve

print('Accuracy =', accuracy_score(y_test, y_predicted))
print('Precision =', precision_score(y_test, y_predicted))
print('Recall =', recall_score(y_test, y_predicted))
print('F1_score =', f1_score(y_test, y_predicted))
print('AUC =', roc_auc_score(y_test, y_proba[:,1]))

fpr, tpr, thresholds = roc_curve(y_test, y_proba[:,1])
```
