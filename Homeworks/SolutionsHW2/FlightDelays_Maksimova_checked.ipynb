### Пробелов нет

```python
comp = pd.get_dummies(data.UniqueCarrier, dummy_na=True, prefix='c') #с - компания-перевозчик
comp.head() #имеем таблицу по компаниям-перевозчикам
```

23 компании, можно оставить в dummy

```python
dataw = pd.concat([data, comp], axis=1)
dataw.head() #совмещаем таблицы w means wide
```

```python
def time_cut(n):
    n=n//100
    return(n)
dataw['Time'] = time_cut(dataw['DepTime'])#новый столбец Time показывает время отправления в часах

dataw = dataw.replace(to_replace=[25], value=[ 1])
dataw = dataw.replace(to_replace=[24], value=[ 0]) #убираем 24 и 25 часов (украдено из первого дз)

dataw = dataw.drop(['UniqueCarrier', 'Origin', 'Dest', 'DepTime'], axis=1) 
dataw.head()
```

```python
len(dataw['Origin to Dest'].unique())
```

4429 явно больше 20, кодируем как частоту. Выкинуть не можем, так как направление может быть свзяано с задержками. 

```python
frequency_encoding = dict(dataw['Origin to Dest'].value_counts()/len(data))

dataw['Origin to Dest'] = dataw['Origin to Dest'].map(frequency_encoding)
dataw.head()
```

Теперь необходимо перевести в dummy дни недели и месяца. Чтобы сократить обьем данных, сгруппируем месяца по сезонам и дни разделим на две категории: пн-чт и пт-вс (предполагаю, они наиболее загруженны)
Так как нам не известно, приведет ли это к улучшению модели, попробуем оба варианта. 

### Комментарий
Очень грамотный подход, группировка действительно иногда может помочь, но запилить два варианта и проверить - самое лучшее решение :)

```python
#Data1 вариант первый: оставляем дамми в большом количестве (как есть)
data1=dataw
data1x1 = pd.get_dummies(data1.DayOfWeek, dummy_na=True, prefix='dow') #dow - день недели
data1x2 = pd.get_dummies(data1.Month, dummy_na=True, prefix='m') #m - месяц
dataw1 = pd.concat([data1, data1x1,data1x2], axis=1)#совмещаем
dataw1 = dataw1.drop(['Month', 'DayOfWeek', 'm_nan', 'dow_nan'], axis=1) #дропаем лишнее
dataw1.head() #получаем первый вариант полной таблицы
```

```python
# Data2 вариант второй: перекодируем дни недели и месяцы
data2=dataw
data2.DayOfWeek = data2.DayOfWeek.map({1: '1', 2: '0',3: '0', 4: '0',5: '0', 6: '1',7: '1'}) 
# так как в США вс = 1, пт, сб и вс считаем самыми загруженными и равными 1
data2.Month = data2.Month.map({1: 'wi', 2: 'wi',3: 'sp', 4: 'sp',5: 'sp', 6: 'su',7: 'su', 8: 'su',9: 'au', 10: 'au',11: 'au', 12: 'wi'})

data2x1 = pd.get_dummies(data2.DayOfWeek, dummy_na=True, prefix='dow') 
data2x2 = pd.get_dummies(data2.Month, dummy_na=True, prefix='m') 
dataw2 = pd.concat([data2, data2x1,data2x2], axis=1)
dataw2 = dataw2.drop(['Month', 'DayOfWeek', 'm_nan', 'dow_nan'], axis=1) 
dataw2.head() #получаем второй вариант полной таблицы
```

Имеем два перекодированных датасета: исходный dataw1 и модернизированный dataw2

# Работаем с Dataw1

```python
dataw1.head()
```

```python
# разбиение 
from sklearn.model_selection import train_test_split
y = dataw1['dep_delayed_15min'].values
X = dataw1.drop('dep_delayed_15min',axis=1).values
dataw1_train, dataw1_test = train_test_split(dataw1, test_size = 0.3) #разбиение выборки


y_train = dataw1_train['dep_delayed_15min']
y_test = dataw1_test['dep_delayed_15min']

X_train = dataw1_train.drop(['dep_delayed_15min'], axis=1)
X_test = dataw1_test.drop(['dep_delayed_15min'], axis=1)
```

```python
# Первая модель - случайный лес

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50) 
rfc.fit(X_train, y_train)
y_predicted = rfc.predict(X_test)
y_proba = rfc.predict_proba(X_test)
```

```python
 # важность факторов
featureImportance = pd.DataFrame({"feature": dataw1.drop('dep_delayed_15min',axis=1).columns, 
                                  "importance": rfc.feature_importances_})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].plot('bar');
```

Больше всего на задержку влияет дистанция, затем идет направление перелетов и день месяца

```python
# Посмотрим на метрики качества

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve

print('Accuracy =', accuracy_score(y_test, y_predicted))
print('Precision =', precision_score(y_test, y_predicted))
print('Recall =', recall_score(y_test, y_predicted))
print('F1_score =', f1_score(y_test, y_predicted))
print('AUC =', roc_auc_score(y_test, y_proba[:,1]))

fpr, tpr, thresholds = roc_curve(y_test, y_proba[:,1])

plt.plot(fpr, tpr, label='ROC curve ')
plt.plot([0, 1], [0, 1])
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.show()
```

```python
# Сделаем дополнительную настройку параметров

from sklearn.model_selection import GridSearchCV

parameters = {
              'n_estimators': [500, 700, 1000, 1500],           
                 
             } 
rfc = RandomForestClassifier(random_state=42)   

gcv = GridSearchCV(rfc, parameters, n_jobs= -1) 
gcv.fit(X, y)
gcv.best_estimator_
```

```python
 #Вторая модель - логистическая регрессия

from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(X_train, y_train)
y_predicted = logit.predict(X_test)
y_proba = logit.predict_proba(X_test)

```

```python
y_mean = np.mean(y_train)                      
y_pred_naive = np.ones(len(y_test)) * y_mean  
y_pred_naive[:1]
```

```python
print('Accuracy =', accuracy_score(y_test, y_predicted))
print('Precision =', precision_score(y_test, y_predicted))
print('Recall =', recall_score(y_test, y_predicted))
print('F1_score =', f1_score(y_test, y_predicted))
print('AUC =', roc_auc_score(y_test, y_proba[:,1]))

fpr, tpr, thresholds = roc_curve(y_test, y_proba[:,1])

plt.plot(fpr, tpr, label='ROC curve ')
plt.plot([0, 1], [0, 1])
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.show()
```
