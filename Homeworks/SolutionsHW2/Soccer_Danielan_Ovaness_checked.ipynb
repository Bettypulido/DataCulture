```python
#there are no nulls

#as there are no categorical variables with less than 20 unique values, I will take the variable 'League'
```

```python
len(soc['league'].unique())
```

```python
frequency_encoding = dict(soc['league'].value_counts()/len(soc))
```

```python
soc['league'] = soc['league'].map(frequency_encoding)
```

```python
soc
```

```python
from sklearn.model_selection import train_test_split
soc_train, soc_test = train_test_split(soc,test_size = 0.3)

y_train = soc_train['RESULT']
y_test = soc_test['RESULT']

X_train = soc_train.drop(['RESULT'], axis=1)
X_test = soc_test.drop(['RESULT'], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logit = LogisticRegression()
```

```python
logit.fit(X_train, y_train)
```

```python
y_predicted = logit.predict(X_test) 
y_proba = logit.predict_proba(X_test)
```

```python
from sklearn import metrics 

print(metrics.classification_report(y_test, y_predicted))
```
