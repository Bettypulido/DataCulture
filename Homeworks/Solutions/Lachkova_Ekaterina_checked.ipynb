### Комментарий к решению
И снова верно

```python
print('Количество уникальных категорий отелей:', data.drop_duplicates(subset = 'categories', keep = False).count().iloc[0])
```

### Комментарий к решению
Однако крутая визуализация

```python
g = sns.pairplot(data[["latitude", "longitude", "postalCode", "reviews.rating"]].dropna(), 
                 hue="reviews.rating", diag_kind="hist")
 
for ax in g.axes.flat: 
    plt.setp(ax.get_xticklabels(), rotation=45)
```

### Комментарий к решению
А вот и карта мира получилась, хотя авторы датасеты уверяли, что у них тут отзывы на американские отели. Придется потом по координатам обрезать только США, чтобы строить модельки :)

```python
plt.figure(figsize = (25,20))
plt.scatter(x=data['longitude'].dropna(), 
                       y=data['latitude'].dropna())
```
