### Комментарий к решению
А вот здесь вы просто посчитали сколько есть квартир с каждым из наборов комнат, чтобы средние посчитать, надо было взять функцию mean():

И тогда сразу интересная зависимость с девятью комнатами видна :)

```python
data.groupby(["Rooms"])[["Price"]].mean().plot.bar();
```

```python
data.groupby(["Rooms"])[["Price"]].count().plot.bar();
```

```python
#Цена резко падает если у недвижимости более 3 комнат.
```

```python
#После группировки мы видим, что у нас отсутствуют данные для 11, 13, 14 и 15 комнат.
#Самое маленькое количество комнат составляет 1, а соме большое 16 . 
#Поэтому среднее значения для этого количества комнат по цене мы найти не сможем.
```

```python
i=1
while i < 11:
    p=int(data.loc[:,['Price']][data.Rooms == i].sum())
    d=int(data.loc[:,['Rooms']][data.Rooms == i].count())
    r=p/d
    print ('Средняя цена для ',i, 'комнат: ', r)
    i=i+1
    
a=int(data.loc[:,['Price']][data.Rooms == 12].sum())
s=int(data.loc[:,['Rooms']][data.Rooms == 12].count())
g=a/s
print ('Средняя цена для ',12, 'комнат: ', g)

a=int(data.loc[:,['Price']][data.Rooms == 16].sum())
s=int(data.loc[:,['Rooms']][data.Rooms == 16].count())
g=a/s
print ('Средняя цена для ',16, 'комнат: ', g)
```
